```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json('data/completeimdb.json')
```

```python
df.isna().sum()/len(df)
```

```python
a = []
for i in list(df['runtime']):
    try :
        a.append(int(i[0]))
    except:
        a.append(0)
```

```python
len(a),a
```

```python
df['runtime'] = a
```

```python
b = []
for i in list(df['runtimes']):
    try :
        b.append(int(i[0]))
    except:
        b.append(0)
```

```python
len(b),b
```

```python
df['runtimes'] = b
```

```python
df['runtime'] = df['runtime'] + df['runtimes']
```

```python
df['runtime'].replace(to_replace = 0,value = np.nan,inplace  =True)
```

```python
df.isna().sum()/len(df)
```

```python
df.drop(['runtimes'],inplace = True,axis = 1)
```

```python
df['rating'].fillna(df['rating'].mean(),inplace=True)
df['vote'].fillna(df['vote'].mean(),inplace=True)
```

```python
df.isna().sum()/len(df)
```

```python
nu = df[df['composer'].isnull()].index.tolist()
```

```python
df.drop(index = nu,inplace = True,axis = 1)
```

```python
df.isna().sum()/len(df),len(df)
```

```python
nu = df[df['language'].isnull()].index.tolist()
df.drop(index = nu,inplace = True,axis = 1)
```

```python
df['runtime'].fillna(df['runtime'].mean(),inplace=True)
```

```python
df.isna().sum()/len(df),len(df)
```

```python
df['year'].fillna(df['year'].mean(),inplace=True)
```

```python
a = []
for i in list(df['language']) :
    b = []
    for j in i :
        if j != 'None':
            b.append(j)
    if len(b) != 0:
        a.append(b)
    else:
        a.append(np.nan)
```

```python
df['language'] = a
```

```python
df.dropna(inplace = True)
```
