## Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def combine_features(row):
    return row['title']+" "+row['year']+" "+row['kind']+" "+row['genre'] + " " + row['country'] + " "+row['language'] + ' ' + row['cast'] + " " + row['director']+" "+row['writer']
```

```python
def get_title_from_index(df,index):
    return df[df['index'] == index]["title"].values[0]
def get_index_from_title(df,title):
    return df[df.title == title]['index'].values[0]
```

```python
def output(df,count_matrix,movie):
    cosine_sim = cosine_similarity(count_matrix)
    movie_index = get_index_from_title(df,movie)
    similar_movies = list(enumerate(cosine_sim[movie_index]))
    sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
    i=0
    out = []
    for element in sorted_similar_movies:
        try:
            out.append(get_title_from_index(df,element[0]))
            i=i+1
            if i>20:
                break
        except:
            continue
    return out
```

```python
def cs_predict(movies,movie):
    df = pd.read_csv('shortened_data.csv')
    df = df.iloc[movies]
    df = df.drop(['Unnamed: 0'],axis = 1)
    df = df.reset_index()
    df = df.drop(['index'],axis = 1)
    df = df.rename_axis('index').reset_index()
    df['year']=df['year'].astype('int')
    features = ['title','year','kind','genre','country','language','cast','director','writer']
    df['genre'] = df['genre'].apply(lambda x : x[1:-1])
    df['country'] = df['country'].apply(lambda x : x[1:-1])
    df['language'] = df['language'].apply(lambda x : x[1:-1])
    df['cast'] = df['cast'].apply(lambda x : x[1:-1])
    df['director'] = df['director'].apply(lambda x : x[1:-1])
    df['writer'] = df['writer'].apply(lambda x : x[1:-1])
    df['year'] = df['year'].astype(str)
    df["combined_features"] = df.apply(combine_features,axis=1)
    cv = CountVectorizer()
    count_matrix = cv.fit_transform(df["combined_features"])
    result = output(df,count_matrix,movie)
    return result
```

```python
def predict(movie,kmeans,fr,fv):
    p = fr[fr['title'] == movie]['index'].iloc[0]
    cl = kmeans.predict(np.reshape(fv[p],(1,-1)))[0]
    if cl == 0 :
        movies = cs_predict(list(fr[(fr['cluster'] == cl) | (fr['cluster'] == cl + 1) | (fr['cluster'] == cl+2)]['index']),movie)
    else:
        movies = cs_predict(list(fr[(fr['cluster'] == cl) | (fr['cluster'] == cl + 1) | (fr['cluster'] == cl-1)]['index']),movie)
    return movies
```
